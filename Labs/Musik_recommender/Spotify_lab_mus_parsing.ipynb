```python
def song_dets(track_with_num):
    album_name = track_with_num['track']["album"]["name"]
    song_artist = []
    for artist in track_with_num['track']["artists"]:
        song_artist.append(artist["name"])
    song_name = track_with_num['track']["name"]
    song_id = track_with_num['track']["id"]
    return song_name, song_id, song_artist, album_name
```

```python
song_dets(tracks[3])
```

```python
song_df = pd.DataFrame(data = song_dets(tracks[0])).T 
song_df.columns = ["song_name", "song_id", "song_artist", "album_name"]
```

```python
for i in new_list:
    tracks = get_playlist_tracks("spotify", i)
    for track in tracks:
        new_song = pd.DataFrame(data = song_dets(track)).T 
        new_song.columns = ["song_name", "song_id", "song_artist", "album_name"]
        song_df = pd.concat([song_df, new_song], axis=0)
    wait_time = randint(1,6)
    print("I will sleep for " + str(wait_time) + " second/s.")
    sleep(wait_time)
```
