```python
import re
songs = []
wrong_words = ['\n\n\t\n\t\n\t\t\n\t\t\t\t\tSongwriter(s):\t\t\n\t\n',  '\n\n\t\n\t\n\t\t\n\t\t\t\t\tProducer(s):\t\t\n\t\n', "\n\n\t\n\t\n\t\t\n\t\t\t\t\tImprint/Promotion Label:\t\t\n\t\n"]
for i in soup.find_all("h3"):
    if i.get_text() not in wrong_words:
        songs.append(i.get_text())
    else:
        pass
songs = songs[4:104]
songs
```

```python
singers = []
for i in range(0,len(soup.select(".c-label"))):
    if "c-label a-no-trucate a-font-primary-s lrv-u-font-size-14" in str(soup.select(".c-label")[i]):
        singers.append(soup.select(".c-label")[i].get_text())
singers
```

```python
import re
```

```python
songs_clean = []
for i in singers:
    songs_clean.append(re.sub(r'[^\w]', ' ', i).strip(" "))
songs_clean = [x.lower() for x in songs_clean]
```

```python
singers_clean = []
for i in songs:
    singers_clean.append(re.sub(r'[^\w]', ' ', i).strip(" "))
singers_clean = [x.lower() for x in singers_clean]
```

```python
top_hund = pd.DataFrame(list(zip(songs_clean, singers_clean)), columns=["Singer", "Song"])
top_hund
```

```python
top_hund.to_csv("top_songs.csv", mode='w',index=False)
```
